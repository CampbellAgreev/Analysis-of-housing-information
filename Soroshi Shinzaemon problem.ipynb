```python
import matplotlib.pyplot as plt

# Code to calculate the total rice grains received on the 100th day

DAYS = 100  # Number of days
initial_rice = 1  # The first day's rice grains

daily_rice = initial_rice  # Rice received on a particular day
total_rice = 0  # Total rice received up to a particular day

daily_rice_list = []  # List to store rice received each day
total_rice_list = []  # List to store cumulative rice count

for _ in range(DAYS):
    daily_rice_list.append(daily_rice)
    total_rice += daily_rice
    total_rice_list.append(total_rice)
    daily_rice *= 2  # Double the rice each day

# Output the total rice received on the 100th day
print("Total rice grains received on the 100th day: {}".format(total_rice))


plt.figure(figsize=(10, 5))

# Daily Rice Received
plt.subplot(1, 2, 1)
plt.title("Daily Rice Received")
plt.xlabel("Day")
plt.ylabel("Rice Grains")
plt.plot(daily_rice_list, color='red')

# Total Rice Accumulated
plt.subplot(1, 2, 2)
plt.title("Total Rice Accumulated")
plt.xlabel("Day")
plt.ylabel("Total Rice Grains")
plt.plot(total_rice_list, color='blue')

plt.tight_layout()
plt.show()

# Explanation of rice growth
print("The number of rice grains increases exponentially. Initially, the increase is small, but after several days, the amount grows to a larger number, making the promised reward unsustainable.")


import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days: int
        Number of days to get rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received each day (length = n_days)
    list_total_grains : list
        List of total rice grains received up to a certain day (length = n_days)
    """
    daily_rice = 1  # The first day's rice grains
    total_rice = 0  # Total rice received up to a particular day

    daily_rice_list = []  # List to store rice received each day
    total_rice_list = []  # List to store cumulative rice count

    for _ in range(n_days):
        daily_rice_list.append(daily_rice)
        total_rice += daily_rice
        total_rice_list.append(total_rice)
        daily_rice *= 2  # Double the rice each day

    return daily_rice_list, total_rice_list

# Number of days to complete the course
n_days = 75
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)

# Output the total rice received on the last day
print("Total rice grains received on day {}: {}".format(n_days, list_total_grains[-1]))


plt.figure(figsize=(10, 5))

# Daily Rice Received
plt.subplot(1, 2, 1)
plt.title("Daily Rice Received")
plt.xlabel("Day")
plt.ylabel("Rice Grains")
plt.plot(list_n_grains, color='indigo')

# Total Rice Accumulated
plt.subplot(1, 2, 2)
plt.title("Total Rice Accumulated")
plt.xlabel("Day")
plt.ylabel("Total Rice Grains")
plt.plot(list_total_grains, color='skyblue')

plt.tight_layout()
plt.show()



import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days: int
        Number of days to get rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received each day (length = n_days)
    list_total_grains : list
        List of total rice grains received up to a certain day (length = n_days)
    """
    daily_rice = 1  # The first day's rice grains
    total_rice = 0  # Total rice received up to a particular day

    daily_rice_list = []  # List to store rice received each day
    total_rice_list = []  # List to store cumulative rice count

    for _ in range(n_days):
        daily_rice_list.append(daily_rice)
        total_rice += daily_rice
        total_rice_list.append(total_rice)
        daily_rice *= 2  # Double the rice each day

    return daily_rice_list, total_rice_list

def calculate_days_of_survival(total_rice, num_people):
    """
    Function to calculate how many days a given amount of rice can sustain a given number of people.
    
    Parameters
    ----------
    total_rice : int
        Total number of rice grains available.
    num_people : int
        Number of people to sustain.
    
    Returns
    -------
    int
        The number of days the rice can sustain the group.
    """
    rice_per_person_per_day = 150000  # Assumption: A person shold consume about 150,000 rice grains per day , that's approximately 3 cups depending on their age and activity level.
    total_daily_consumption = num_people * rice_per_person_per_day
    return total_rice // total_daily_consumption  # Integer division to get whole days


n_days = 75
num_people = 40  # Example class size
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)
total_rice = list_total_grains[-1]

days_of_survival = calculate_days_of_survival(total_rice, num_people)
print("Total rice grains received on day {}: {}".format(n_days, total_rice))
print("With {} people, the rice can sustain them for {} days.".format(num_people, days_of_survival))


plt.figure(figsize=(10, 5))

# Daily Rice Received
plt.subplot(1, 2, 1)
plt.title("Daily Rice Received")
plt.xlabel("Day")
plt.ylabel("Rice Grains")
plt.plot(list_n_grains, color='orange')

# Total Rice Accumulated
plt.subplot(1, 2, 2)
plt.title("Total Rice Accumulated")
plt.xlabel("Day")
plt.ylabel("Total Rice Grains")
plt.plot(list_total_grains, color='skyblue')

plt.tight_layout()
plt.show()


```
