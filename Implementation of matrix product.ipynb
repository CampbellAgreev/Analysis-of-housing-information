A= [(1 2 3), (4 -5 6), (7 8 -9)] 
B= [(0 2 1), (0 2 -8), (2 9 -1)]

C=A×B
compute:
C_ij = ∑_K=1 to 3 (A_ij *B_kj)

Where C_ij is element at row i, column j

First row multiplication
C_11=(−1×0)+(2×0)+(3×2)=0+0+6=6
C_12=(−1×2)+(2×2)+(3×9)=−2+4+27=29
C_13=(−1×1)+(2×−8)+(3×−1)=−1−16−3=−20

Second row multiplication
C_21=(4×0)+(−5×0)+(6×2)=0+0+12=12
C_22=(4×2)+(−5×2)+(6×9)=8−10+54=52
C_23=(4×1)+(−5×−8)+(6×−1)=4+40−6=38

Third row multiplication
C_31=(7×0)+(8×0)+(−9×2)=0+0−18=−18
C_32=(7×2)+(8×2)+(−9×9)=14+16−81=−51
C_33=(7×1)+(8×−8)+(−9×−1)=7−64+9=−48
